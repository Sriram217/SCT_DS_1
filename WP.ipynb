```python
mydata = pd.read_csv('batnew.csv')
Country = mydata['Country']
Population = mydata['Population']
```

```python
fig = plt.figure()
myaxes = fig.add_axes([15.1,15.1,15.6,15.6])
myaxes.bar(Country,Population)
myaxes.set_xlabel('Countries', fontsize=140)
myaxes.set_ylabel('Population', fontsize=140)
myaxes.set_title('World Population', fontsize=200)
plt.show()
```
